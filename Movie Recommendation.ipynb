```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/movies.csv')
df.head()
```

```python
df.shape
```

```python
df_selected = ['genres','keywords','tagline','cast','director']
print(df_selected)
```

```python
for feature in df_selected:
  df[feature]=df[feature].fillna('')
```

```python
df_new = df['genres']+''+df['keywords']+''+df['tagline']+''+df['cast']+''+df['director']
print(df_new)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(df_new)
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
movie_name = input('Enter movie name: ')
```

```python
movie_titles = df['title'].tolist()
print(movie_titles)
```

```python
find_match = difflib.get_close_matches(movie_name,movie_titles)
print(find_match)
```

```python
close_match = find_match[0]
index_movie = df[df.title==close_match]['index'].values[0]
print(index_movie)
```

```python
similarity_score = list(enumerate(similarity[index_movie]))
```

```python
sorted_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_movies)
```

```python
print('Movies Suggessted for you!---')

i=1

for movie in sorted_movies:
  index = movie[0]
  title_from_index = df[df.index==index]['title'].values[0]
  if(i<10):
    print(i,'.',title_from_index)
    i+=1
```
