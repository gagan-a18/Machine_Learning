```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/mail_data.csv')
df.head()
```

```python
df.shape
```

```python
df = df.where((pd.notnull(df)),'')
df.head()
```

```python
df.replace({'Category':{'ham':1,'spam':0}},inplace=True)
df.head()
```

```python
X = df['Message']
y = df['Category']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_new = feature_extraction.fit_transform(X_train)
X_test_new = feature_extraction.transform(X_test)

y_train_new = y_train.astype('int')
y_test_new = y_test.astype('int')
```

```python
print(X_train_new)
```

```python
model = LogisticRegression()
model.fit(X_train_new,y_train_new)
```

```python
y_pred = model.predict(X_test_new)
a = accuracy_score(y_pred,y_test_new)
print(a)
```
